```python
import spacy
import en_core_web_sm
import spacy.attrs
from collections import Counter
```

```python
nlp = spacy.load("en")
nlp = en_core_web_sm.load()
```

```python
article = open('article.txt')
```

```python
text = article.read()
```

```python
print(text)
```

```python
text = nlp(text)
```

```python
print("Word"+" " * 11 + "WORD_POS"+ " "* 7+ "WORD_TAG")
print(" ")
for word in text:
    print(f"{word.text:{15}}{word.pos_:{15}}{word.tag_:{15}}")
```

```python
words = [word.text for word in text if word.is_stop != True and word.is_punct != True]
word_freq = Counter(words)
words_frequency = word_freq.most_common(500)
print (words_frequency)
```
